```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

lag = 50
ax1 = plt.subplot(2, 1, 1)
plot_acf(noise[0,:], lags = lag, ax=ax1, title='ACF for Prey Population ($X_t$)')
ax1.set_xlabel('Lag')
ax1.set_ylabel('Autocorrelation')

ax2 = plt.subplot(2, 1, 2)
plot_pacf(noise[0,:], lags = lag, ax=ax2, title='PACF for Prey Population ($X_t$)', method='ywm') # 'ywm' or 'ols'
ax2.set_xlabel('Lag')
ax2.set_ylabel('Partial Autocorrelation')

plt.tight_layout()
plt.show()
```

```python
from scipy.signal import periodogram

# --- Parameters (Use the same data as before) ---
# N is the number of data points
N = len(noise[0,:])
# Sample spacing (dt) is the time difference between consecutive points.
# Since we used np.linspace(0, 500, 500) previously, the spacing is 500/500 = 1.0
SAMPLE_SPACING = 1.0 # The 'd' parameter in np.fft.fftfreq

def find_dominant_period(data, N, dt):
    """Calculates PSD using periodogram and finds the dominant period."""

    # 1. Calculate the Power Spectral Density (PSD)
    # fs = 1/dt is the sampling frequency
    frequencies, power = periodogram(data, fs=1/dt, scaling='density')

    # We are interested only in positive frequencies (f > 0)
    positive_freq_indices = frequencies > 0
    freqs = frequencies[positive_freq_indices]
    power = power[positive_freq_indices]

    # 2. Find the index corresponding to the maximum power
    dominant_index = np.argmax(power)
    dominant_freq = freqs[dominant_index]

    # 3. Calculate the dominant period (Period = 1 / Frequency)
    if dominant_freq > 0:
        dominant_period = 1 / dominant_freq
    else:
        dominant_period = np.nan

    return freqs, power, dominant_period

# --- Analysis for Prey Population ---
prey_freqs, prey_power, prey_period = find_dominant_period(noise[0,:], N, SAMPLE_SPACING)

print(f"Prey Dominant Cycle Length: {prey_period:.2f} time steps")


plt.figure(figsize=(12,4))
plt.plot(prey_freqs, prey_power, color='blue', alpha=0.7)
plt.axvline(x=1/prey_period, color='red', linestyle='--', label=f'Dominant Freq (1/Period = {1/prey_period:.3f})')
plt.title('Periodogram for Prey Population ($X_t$)')
plt.xlabel('Frequency (cycles/time step)')
plt.ylabel('Power Spectral Density (PSD)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
